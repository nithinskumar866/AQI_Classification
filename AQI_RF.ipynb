```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
dataset['Country'] = label_encoder.fit_transform(dataset['Country'])
dataset['City'] = label_encoder.fit_transform(dataset['City'])
dataset['AQI Value'] = label_encoder.fit_transform(dataset['AQI Value'])
dataset['AQI Category'] = label_encoder.fit_transform(dataset['AQI Category'])
dataset['Ozone AQI Value'] = label_encoder.fit_transform(dataset['Ozone AQI Value'])
dataset['Ozone AQI Category'] = label_encoder.fit_transform(dataset['Ozone AQI Category'])
dataset['NO2 AQI Value'] = label_encoder.fit_transform(dataset['NO2 AQI Value'])
dataset['NO2 AQI Category'] = label_encoder.fit_transform(dataset['NO2 AQI Category'])
dataset['PM2.5 AQI Value'] = label_encoder.fit_transform(dataset['PM2.5 AQI Value'])
dataset['PM2.5 AQI Category'] = label_encoder.fit_transform(dataset['PM2.5 AQI Category'])
dataset['lat'] = label_encoder.fit_transform(dataset['lat'])
dataset['lng'] = label_encoder.fit_transform(dataset['lng'])
```

```python
columns_to_drop = ['Country', 'City', 'NO2 AQI Value', 'NO2 AQI Category']
dataset = dataset.drop(columns_to_drop, axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=dataset,x='lng',y='lat',hue='AQI Category',palette='Set1')
plt.title('AQI Categories around the world',fontsize=20)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
X = dataset[['AQI Value', 'CO AQI Value', 'Ozone AQI Value', 'PM2.5 AQI Value', 'lat', 'lng']]
Y = dataset['AQI Category']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
print('x_train: ', x_train.shape)
print('x_test: ', x_test.shape)
print('y_train: ', y_train.shape)
print('y_test: ', y_test.shape)
```

```python
from sklearnex.ensemble import RandomForestClassifier
from sklearnex import patch_sklearn
patch_sklearn()
```

```python
score = []
for i in range(1,51):
    clf = RandomForestClassifier(criterion = 'entropy', n_estimators = i)
    clf.fit(x_train,y_train)
    scores = clf.score(x_test,y_test)
    score.append(scores)
```

```python
score_rate = pd.DataFrame(data = score, columns = ['Score'])
score_rate.insert(loc=0, column = 'n_estimators', value=range(1,51))
```

```python
model = RandomForestClassifier(criterion='entropy', n_estimators=30)
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)*100
```

```python
predict = model.predict(x_test)
predict
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy Score: ', accuracy_score(y_test, predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, predict)
cm
```

```python
# visualize confusion metrix

ax = sns.heatmap(cm, annot=True, cmap='Blues')

ax.set_title('Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

plt.show()
```

```python
ax = sns.heatmap(cm/np.sum(cm), annot=True,  fmt='.2%', cmap='Blues')

ax.set_title('Confusion Matrix with labels\n\n');
ax.set_xlabel('\nPredicted Values')
ax.set_ylabel('Actual Values ');

plt.show()
```
